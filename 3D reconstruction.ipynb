```python
database_path = 'data.db'
image_path = 'Challenge/png/Scene2'
out = 'Challenge/png/Scene2'
```

```python

subprocess.run(["colmap", "feature_extractor", "--database_path", database_path, "--image_path", image_path, "--SiftExtraction.max_num_features", "50000"])
subprocess.run(["colmap", "exhaustive_matcher", "--database_path", database_path, '--SiftMatching.use_gpu', '0'])
subprocess.run(["colmap", "mapper", "--database_path", database_path, "--image_path", image_path, "--output_path", out,"--Mapper.num_threads","8"])
```
